```python
import pandas as pd
df = pd.read_csv('creditcard.csv')
df
```

```python
df['Class'].value_counts()
```

```python
from sklearn.preprocessing import RobustScaler
new_df = df.copy()
new_df['Amount'] = RobustScaler().fit_transform(new_df['Amount'].to_numpy().reshape(-1, 1))
time = new_df['Time']
new_df['Time'] = (time - time.min()) / (time.max() - time.min())
new_df
```

```python
new_df = new_df.sample(frac=1, random_state=1)
new_df
```

```python
train, test, val = new_df[:int(len(new_df)*0.8)], new_df[int(len(new_df)*0.8):int(len(new_df)*0.9)], new_df[int(len(new_df)*0.9):]
train['Class'].value_counts(), test['Class'].value_counts(), val['Class'].value_counts()
```

```python
train_np, test_np, val_np = train.to_numpy(), test.to_numpy(), val.to_numpy()
train_np.shape, test_np.shape, val_np.shape
```

```python
x_train, y_train = train_np[:, :-1], train_np[:, -1]
x_test, y_test = test_np[:, :-1], test_np[:, -1]
x_val, y_val = val_np[:, :-1], val_np[:, -1]
x_train.shape, y_train.shape, x_test.shape, y_test.shape, x_val.shape, y_val.shape
```

```python
from sklearn.ensemble import IsolationForest
# All features including PCA components, Time, and Amount
X = new_df[[f"V{i}" for i in range(1, 29)] + ['Time', 'Amount']].values
y = new_df['Class'].values
```

```python
# Remove any NaNs just in case
y_clean = y[~pd.isna(y)]
X_clean = X[~pd.isna(y)]
```

```python
contamination_rate = y_clean.mean()  # should now be numeric
print(f"Contamination rate: {contamination_rate:.6f}")
```

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(
    n_estimators=100,
    max_samples='auto',
    contamination=contamination_rate,  # numeric value now
    random_state=42
)

iso_forest.fit(X_clean)
```

```python
# Remove rows with NaN in y
mask = ~np.isnan(y)
X_clean = X[mask]
y_clean = y[mask]

# Predict using IsolationForest
pred = iso_forest.predict(X_clean)  # returns 1 (normal) or -1 (anomaly)
pred_labels = np.where(pred == -1, 1, 0)  # Convert -1 -> 1 (Fraud), 1 -> 0 (Not Fraud)

# Evaluate
from sklearn.metrics import classification_report
print(classification_report(y_clean, pred_labels, target_names=['Not Fraud', 'Fraud']))
```
